## Circuit 1: Single Qubit Superposition

### What is a qubit?

A ==qubit== (quantum bit) is the quantum version of a classical bit. Unlike a classical bit (which is 0 or 1), a qubit can exist in ==superposition==: a combination of 0 and 1 simultaneously.

一個==量子位元==(qubit) 是古典位元的量子版本。不像古典位元（0 或 1），量子位元可以存在於==疊加態==：0 和 1 的組合。

### The Hadamard Gate

The ==Hadamard gate== (H) creates superposition. If a qubit starts in state |0⟩, applying H puts it into superposition:

```
|0⟩ ──H── (|0⟩ + |1⟩) / √2
```

Meaning: the qubit is now equally likely to be measured as 0 or 1 (50/50 probability).

==Hadamard gate== (H) 創造疊加態。如果量子位元從態 |0⟩ 開始，應用 H 將它放入疊加態。

### The Circuit: H on a Single Qubit

Circuit 1：具體例子
想像你有一個魔法硬幣。

古典硬幣： 正面或反面。一旦翻出來，就確定了。
量子硬幣（魔法）： 在你看它之前，它同時是正面和反面。一旦你看，它才決定成正面或反面。

我們在做什麼

````python
# Create a quantum circuit with 1 qubit and 1 classical bit (for measurement)建立 1 個量子位元 + 1 個古典位元（用來測量）
qc1 = QuantumCircuit(1, 1)

# Apply Hadamard gate to create superposition在量子位元 0 上套用 Hadamard 門（魔法）
#    把 |0⟩ 變成「同時是 0 和 1」
qc1.h(0)

# 3. 測量（我們看它）
#    它被迫選擇：0 或 1
qc1.measure(0, 0)

# Display the circuit顯示電路圖
print("Circuit 1: Single Qubit Superposition")
print(qc1.draw(output='text'))
print("\nCircuit diagram saved as: 01_circuit1_single_qubit.png")
````

````python
# Run the circuit on simulator在模擬器上執行 1000 次
simulator = AerSimulator()
job1 = simulator.run(qc1, shots=1000)  # Run 1000 times to see the distribution
counts1 = job1.result().get_counts()

print("\nMeasurement results (1000 shots):")
print(counts1)
print(f"\nProbability of measuring 0: {counts1.get('0', 0) / 1000:.1%}")
print(f"Probability of measuring 1: {counts1.get('1', 0) / 1000:.1%}")
print("\nExpected: ~50% each (they should be roughly equal)")
````

為什麼這很重要？
如果古典電腦做這個實驗，結果會是：

每次都是 0（或每次都是 1）
但量子電腦的結果是：

大約一半是 0，一半是 1
這證明了在測量之前，它同時是 0 和 1

````python
# Visualise the measurement distribution
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
plot_histogram(counts1, ax=ax, title="Circuit 1: Measurement outcomes (H gate)")
plt.tight_layout()
plt.savefig('figures/01_circuit1_histogram.png', dpi=150, bbox_inches='tight')
plt.show()

print("Histogram saved to: figures/01_circuit1_histogram.png")
````

### What just happened?

1. We created a qubit in state |0⟩ (the default starting state)
2. Applied a Hadamard gate, putting it into superposition
3. Measured the qubit, which collapsed it to 0 or 1
4. Ran this 1000 times to see the probability distribution
5. Result: ~500 zeros and ~500 ones (50/50 split)

**Key insight:** Superposition only exists before measurement. Once we measure, it collapses to a definite value.

我們創建一個量子位元，應用 Hadamard gate，測量結果崩坍為 0 或 1。重複 1000 次顯示 50/50 機率。

## Circuit 2: Bell State (Entangled Qubits)

### What is entanglement?

==Entanglement== means two or more qubits are correlated in a way impossible classically. If you measure one qubit, it instantly affects the other, even though they're not connected.

==糾纏==意味著兩個或多個量子位元以古典上不可能的方式相關聯。測量一個量子位元會立即影響另一個。

### The CNOT Gate

A ==CNOT== (Controlled-NOT) gate entangles two qubits. It has a "control" qubit and a "target" qubit:

- If control qubit is |0⟩, do nothing
- If control qubit is |1⟩, flip the target qubit

### Creating a Bell State

The circuit:
```
q0 ──H──●──
        │
q1 ─────X──
```

This creates a ==Bell state==: (|00⟩ + |11⟩) / √2

Meaning: q0 and q1 are perfectly correlated. Measure q0 as 0 and q1 will always be 0. Measure q0 as 1 and q1 will always be 1.



大圖景：我們在做什麼？
目標： 證明兩個量子位元能以「古典上不可能的方式」連接在一起。
這叫 糾纏 (Entanglement)。

簡單類比
想像你有 兩個魔法硬幣 A 和 B。

古典情況：

硬幣 A：正面或反面（隨機）
硬幣 B：正面或反面（隨機，獨立）
結果：四種可能都能出現（正正、正反、反正、反反）
糾纏情況（魔法）：

我們施展一個咒語，讓 A 和 B 「同步」
如果 A 是正面，B 必定是正面
如果 A 是反面，B 必定是反面
結果：只有兩種可能（正正、反反），永遠不會出現正反或反正
這就是糾纏。

````python
# Create a quantum circuit with 2 qubits and 2 classical bits 建立 2 個量子位元 + 2 個古典位元（用來測量）
qc2 = QuantumCircuit(2, 2)

# Apply Hadamard to first qubit (creates superposition)在第一個量子位元 (q0) 套用 Hadamard: 把它變成「同時是 0 和 1」
qc2.h(0)

# Apply CNOT: control=0, target=1 (creates entanglement)（這是咒語）
#    控制位元：q0
#    目標位元：q1
#    規則：如果 q0 是 |1⟩，就翻轉 q1
qc2.cx(0, 1)  # cx = CNOT (Controlled-NOT)

# Measure both qubits測量兩個位元
qc2.measure([0, 1], [0, 1])

# Display the circuit顯示電路圖
print("Circuit 2: Bell State (Entangled Qubits)")
print(qc2.draw(output='text'))
````

````python
# Run the circuit執行 1000 次
job2 = simulator.run(qc2, shots=1000)
counts2 = job2.result().get_counts()

print("\nMeasurement results (1000 shots):")
print(counts2)
print(f"\nProbability of |00⟩: {counts2.get('00', 0) / 1000:.1%}")
print(f"Probability of |11⟩: {counts2.get('11', 0) / 1000:.1%}")
print(f"Probability of |01⟩: {counts2.get('01', 0) / 1000:.1%}")
print(f"Probability of |10⟩: {counts2.get('10', 0) / 1000:.1%}")
print("\nExpected: ~50% for |00⟩ and ~50% for |11⟩, ~0% for |01⟩ and |10⟩")
print("Notice: we NEVER see |01⟩ or |10⟩. The qubits are perfectly correlated.")
````

為什麼這很重要？
古典硬幣的預期：

正正：25%
正反：25%
反正：25%
反反：25%
量子糾纏的實際結果：

正正：50%
正反：0%     ← 消失了！
反正：0%     ← 消失了！
反反：50%
這不能用古典機率解釋。 兩個位元不是獨立的；測量 q0 會決定 q1 的值。

簡單總結
Circuit 1： 一個位元「同時是 0 和 1」 ✓

Circuit 2： 兩個位元「同時 0-0 和 1-1，永遠不是 0-1 或 1-0」✓

````python
# Visualise
# Add the missing states (|01⟩ and |10⟩) to show they are 0%
counts2_full = {'00': counts2.get('00', 0),
                '01': counts2.get('01', 0),
                '10': counts2.get('10', 0),
                '11': counts2.get('11', 0)}

fig, ax = plt.subplots(1, 1, figsize=(8, 5))
plot_histogram(counts2_full, ax=ax, title="Circuit 2: Bell State (Entangled)\n|00⟩ and |11⟩ only, never |01⟩ or |10⟩")
plt.tight_layout()
plt.savefig('figures/02_circuit2_bell_state.png', dpi=150, bbox_inches='tight')
plt.show()

print("Histogram saved to: figures/02_circuit2_bell_state.png")
````

### What just happened?

1. Hadamard gate on q0 created superposition: (|0⟩ + |1⟩) / √2
2. CNOT gate entangled q0 and q1: (|00⟩ + |11⟩) / √2
3. Measuring q0 as 0 guarantees q1 is 0. Measuring q0 as 1 guarantees q1 is 1.
4. The states |01⟩ and |10⟩ are impossible—they never appear, even after 1000 shots.

**Key insight:** This is non-classical behaviour. Classically, if we don't know q0's value before measurement, we can't predict q1 either. But quantum entanglement makes them perfectly correlated.

Hadamard gate 對 q0 創造疊加態，CNOT gate 糾纏 q0 和 q1。測量結果只有 |00⟩ 和 |11⟩，永遠不會出現 |01⟩ 或 |10⟩。

## Circuit 3: GHZ State (3-Qubit Entanglement)

### Going Deeper: 3-Qubit Entanglement

The GHZ state entangles three qubits: (|000⟩ + |111⟩) / √2

This demonstrates that entanglement scales. Measuring q0 as 0 guarantees q1 and q2 are also 0.

### The Circuit

大圖景：我們在做什麼？
目標： 證明糾纏可以擴展到三個量子位元（而不只是兩個）。

這會讓你看到：為什麼量子電腦隨著位元增多會變得更強大。

簡單類比
回想 Circuit 2：兩個魔法硬幣 A 和 B 同步

結果：只有兩種可能（正正、反反）
現在加第三個硬幣 C：三個魔法硬幣 A、B、C 全部同步

A 是正面 → B 必定正面 → C 必定正面
A 是反面 → B 必定反面 → C 必定反面
結果：只有兩種可能（正正正、反反反）
永遠不會出現混合的狀態，例如正正反或反反正。

````python
# Create a quantum circuit with 3 qubits and 3 classical bits 建立 3 個量子位元 + 3 個古典位元
qc3 = QuantumCircuit(3, 3)

# Apply Hadamard to first qubit把它變成「同時是 0 和 1」
qc3.h(0)

# Entangle q1 with q0  CNOT：q0 控制 q1（q0 和 q1 同步）
qc3.cx(0, 1)

# Entangle q2 with q0   CNOT：q0 控制 q2（q0 和 q2 同步）
qc3.cx(0, 2)

# 現在三個位元全部糾纏在一起 
# Measure all three qubits  測量全部三個位元
qc3.measure([0, 1, 2], [0, 1, 2])

# Display the circuit 
print("Circuit 3: GHZ State (3-Qubit Entanglement)")
print(qc3.draw(output='text'))
````

````python
# Run the circuit
job3 = simulator.run(qc3, shots=1000)
counts3 = job3.result().get_counts()

print("\nMeasurement results (1000 shots):")
for state, count in sorted(counts3.items()):
    print(f"  |{state}⟩: {count:4d} counts ({count/1000:.1%})")

print(f"\nProbability of |000⟩: {counts3.get('000', 0) / 1000:.1%}")
print(f"Probability of |111⟩: {counts3.get('111', 0) / 1000:.1%}")
print(f"Probability of all others: {(1000 - counts3.get('000', 0) - counts3.get('111', 0)) / 1000:.1%}")
print("\nExpected: ~50% |000⟩ and ~50% |111⟩, ~0% everything else")
````

結果例子：

|000⟩: 492 次 (49.2%)  ✓ 全 0
|001⟩: 0 次
|010⟩: 0 次
|011⟩: 0 次
|100⟩: 0 次
|101⟩: 0 次
|110⟩: 0 次
|111⟩: 508 次 (50.8%)  ✓ 全 1

為什麼這很重要？

古典三硬幣的預期：
正正正：12.5%
正正反：12.5%
正反正：12.5%
... (8 種都各 12.5%)

量子糾纏的實際結果：
正正正：50%
正正反：0%     ← 消失了
正反正：0%     ← 消失了
... (6 種都是 0%)
反反反：50%

關鍵洞察
位元數	只出現的狀態	範例
1 個	2 種（	0⟩ 或
2 個	2 種（	00⟩ 或
3 個	2 種（	000⟩ 或
N 個	2 種（全 0 或全 1）	GHZ State
這叫 GHZ State（Greenberger-Horne-Zeilinger state），是最大的糾纏狀態。

簡單總結
Circuit 1： 一個位元同時是 0 和 1 ✓

Circuit 2： 兩個位元同時 0-0 和 1-1 ✓

Circuit 3： 三個位元同時 0-0-0 和 1-1-1 ✓

Next Level： 如果加 100 個位元，全部糾纏，結果只有兩種：全 0 或全 1。

這就是量子電腦為什麼強大。古典電腦需要檢查 2^100 種可能性。量子電腦可以同時測試它們。

````python
# Visualise
# Add all 8 possible states to show which ones are 0%
counts3_full = {'000': counts3.get('000', 0),
                '001': counts3.get('001', 0),
                '010': counts3.get('010', 0),
                '011': counts3.get('011', 0),
                '100': counts3.get('100', 0),
                '101': counts3.get('101', 0),
                '110': counts3.get('110', 0),
                '111': counts3.get('111', 0)}

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
plot_histogram(counts3_full, ax=ax, title="Circuit 3: GHZ State (3-Qubit Entanglement)\n|000⟩ and |111⟩ only, all others are 0%")
plt.tight_layout()
plt.savefig('figures/03_circuit3_ghz_state.png', dpi=150, bbox_inches='tight')
plt.show()

print("Histogram saved to: figures/03_circuit3_ghz_state.png")
````

### What just happened?

1. Hadamard on q0 creates superposition
2. Two CNOT gates (from q0 to q1, then q0 to q2) entangle all three qubits
3. Result: the GHZ state (|000⟩ + |111⟩) / √2
4. Measuring q0 determines q1 and q2 simultaneously
5. We see only |000⟩ and |111⟩, never intermediate states like |001⟩ or |110⟩

**Key insight:** Adding more qubits doesn't make them independent—instead, entanglement scales. This is why quantum computers are powerful: many qubits can be entangled and processed in parallel.

糾纏在三個量子位元上擴展。測量 q0 同時決定 q1 和 q2。我們只看到 |000⟩ 和 |111⟩，永遠不會看到中間態。

## Summary: Key Insights | 摘要：核心見解

### Superposition (疊加態)
A qubit can be 0 and 1 simultaneously until measured. The Hadamard gate creates equal superposition.
- Single qubit: measured as 0 or 1 with 50% each
- Captured in Circuit 1

### Entanglement (糾纏)
Multiple qubits can be correlated in non-classical ways. Measuring one qubit instantly determines the others.
- Two qubits: always measured as |00⟩ or |11⟩, never |01⟩ or |10⟩
- Three qubits: always measured as |000⟩ or |111⟩
- Captured in Circuits 2 and 3

### Measurement (測量)
Measurement collapses superposition. Before measurement, the qubit is in superposition; after measurement, it's definite.

### Why This Matters for Quantum Algorithms
Superposition + entanglement + interference allow quantum computers to:
1. Test many possibilities simultaneously (superposition)
2. Correlate information across qubits (entanglement)
3. Amplify correct answers and cancel wrong ones (interference)

This is how Amplitude Estimation (Week 2) will work: we'll encode a probability as a quantum amplitude, use entanglement to amplify it, and use phase estimation to extract the answer.

### 三個核心概念

| 概念 | 解釋 | 證據（Circuit） |
|---|---|---|
| **疊加態** | 一個位元同時是 0 和 1 | Circuit 1：50% 是 0，50% 是 1 |
| **糾纏** | 多個位元被迫同步 | Circuit 2：只看到 00 或 11，永不 01 或 10 |
| **擴展性** | 糾纏可以加到更多位元 | Circuit 3：只看到 000 或 111，其他都是 0% |

## 你實際寫了什麼 (Code)

| 步驟 | 代碼 | 做什麼 |
|---|---|---|
| 1 | `qc = QuantumCircuit(3, 3)` | 建立 3 個量子位元 + 3 個古典位元 |
| 2 | `qc.h(0)` | Hadamard：q0 變成疊加態 |
| 3 | `qc.cx(0, 1)` | CNOT：q0 控制 q1（糾纏） |
| 4 | `qc.cx(0, 2)` | CNOT：q0 控制 q2（糾纏） |
| 5 | `qc.measure([0, 1, 2], [0, 1, 2])` | 測量三個位元 |
| 6 | `simulator.run(qc, shots=1000)` | 執行 1000 次 |
| 7 | `plot_histogram(counts)` | 看結果 |
